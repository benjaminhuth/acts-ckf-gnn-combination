```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import random

from gnn4itk_tools.detector_plotter import DetectorPlotter
```

```python
plotter = DetectorPlotter("../output/detectors.csv", r_max=200, abs_z_max=1500)
```

```python
prototracks = pd.read_csv("../output/event000000000-prototracks.csv")
prototracks["r"] = np.hypot(prototracks.x, prototracks.y)
```

```python
tracks = [ g[1] for g in prototracks.groupby("trackId") ]
len(tracks)
```

```python
_ = plt.hist([ len(t) for t in tracks ], bins=np.linspace(2.5,20,20))
```

```python
fig, ax = plotter.get_fig_ax(figsize=(15,7))

for track in random.sample(tracks, 10):
    track = track.sort_values('r').copy()
    ax[0].plot(track.z, track.r)
    ax[1].plot(track.x, track.y)
```
